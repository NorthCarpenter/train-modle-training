```python
# Import Block

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the Merged Train Data Sheet

clean_train_CSV_path = '/Users/NorthCarpenter/Documents/Sort/Employment/Making Employment/Train Code/cleanAllTrainData.csv'

train_data = pd.read_csv(clean_train_CSV_path)
```

```python
# Summary Stats
train_data.describe()
```

```python
# Time Organized Visualisations

train_data.plot.line(x = 'Date Time', y = 'Delay')
plt.show()
```

```python
train_data.head()
train_data.dtypes
```

```python
# Density Plots

sns.set_style('whitegrid')
inlier_data = train_data[train_data['Delay'] <= 100]
inlier_data = inlier_data[inlier_data['Delay'] >= -50]
sns.kdeplot(inlier_data, x = 'Delay', bw = 0.05 )
plt.show()
```

```python
count_within = ((train_data['Delay'] >= -60) & (train_data['Delay'] <= 60)).sum()

print(len(train_data))

print(count_within)

print(count_within/len(train_data))
```
